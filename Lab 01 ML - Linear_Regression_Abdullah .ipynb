## Preprocess the Data

#### 2. Feature engineering

1. Feature scaling
2. Aggregation
3. One hot coding

### Aggregat
Aggregating  categorical feature

```python
categorical_features = ['Team',  'Position']
categorical_features
```

```python
df.shape
```

### One hot coding

```python
df = pd.get_dummies(df, columns=categorical_features)
```

```python
df.shape
```

```python
df.head()
```

### Feature selection
selecting  features  that make more impact on true value

```python
Current_Value_correlation = df.corr()
print(Current_Value_correlation['Current_Value'].sort_values(ascending=False))
```

```python
threshold = 0.12 
selected_features = Current_Value_correlation[abs(Current_Value_correlation['Current_Value']) > threshold]['Current_Value'].index
selected_features
```

```python
selected_features = ['Appearances', 'Minutes_Played', 'Award', 'Current_Value',
       'Highest_Value', 'Team_Leeds United', 'Team_Southampton FC']
```

```python
df = df[selected_features]
df.head()
```

### Prepare train and test data
Prepare and divide data into training data and test data.

```python
X = df.drop(['Current_Value'], axis=1)
y = df['Current_Value']
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=42)
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X.shape
```

### Buliding the Model
Bulid a Linear Regression Model

```python
model = LinearRegression()
```

```python
ridge_model = Ridge()  
```

```python
lasso_model = Lasso() 
```

```python
#hyperparameters
param_grid = {
    'alpha': [0, 0.0000003, 0.0002, 0.001]  
}
grid_search = GridSearchCV(estimator=ridge_model,
                           param_grid=param_grid,
                           cv=5,
                           scoring='neg_mean_squared_error', 
                           verbose=1)

grid_search_lasso = GridSearchCV(estimator=lasso_model,
                           param_grid=param_grid,
                           cv=5,
                           scoring='neg_mean_squared_error', 
                           verbose=1)
```

### Training the model by fit train data

```python

model.fit(X_train_scaled, y_train)
```

#### Test the Model
##### Make a Predict 

```python
y_pred = model.predict(X_test_scaled)
```
